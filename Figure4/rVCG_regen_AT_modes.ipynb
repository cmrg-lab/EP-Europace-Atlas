```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from sklearn.decomposition import PCA
import seaborn as sns; sns.set()
from IPython.display import display

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

from scipy.signal import resample
from scipy import stats
import glob
import sys
import joblib
sys.path.insert(-1,'../Tools')
from data_modification import calc_cost_patient, normalize_patient, trim_tat_patient, resample_patient, scale_at_patient
from data_analysis import df_vcg_vector, plot_vcgs, calc_cost, calc_cost_rms, calc_cost_rms_norm
from data_parser_new import parse
```

# Load simulation data

```python
data_path = '../Data/New_Simulations/'
patients = [
            parse(initial_path=data_path, patient_number=patient_no, at_map = True, verbose = False) 
        #for patient_no in [0]
        for patient_no in range(0,8)
]
```

```python
num_samples = 100
patients = [normalize_patient(patient) for patient in patients]
patients = [trim_tat_patient(patient) for patient in patients]
patients = [resample_patient(patient, length=num_samples) for patient in patients]
patients = [scale_at_patient(patient, length=num_samples) for patient in patients]
```

# Gather optimal VCGs

```python
# copied from Generate_rVCG_AT_modes.ipynb
pt_opt_theta = {'BiV7': 54, 'BiV8': 54, 'BiV6': 70, 'BiV1': 144, 'BiV4': 59, 'BiV3': 37, 'BiV2': 55, 'BiV5': 88}
```

```python
pt_opt_vcg_theta = pd.DataFrame()

for patient in patients:
    pt = patient['pt_id']
    
    x = patient['vcg_model'][pt_opt_theta[pt]] 
    pt_opt_vcg_theta[pt] = pd.concat([x['VCGx'],x['VCGy'],x['VCGz']],axis=0,ignore_index=True)
```

# Non-normalized AT map VCGs

```python
#Load patient data
vcg_files = glob.glob('../Data/rVCG_regen_AT_modes/raw/VCG_*.txt')
vcg_files.sort()

vcg_df = pd.DataFrame()

for f in vcg_files:
    vcg = pd.read_csv(f, header=None, usecols=[2, 3, 4], names=['VCGx', 'VCGy', 'VCGz'], sep='\t', index_col=False)
    fname = f.split('/')[-1].split('_',1)[-1].strip('.txt')
    vcg_df[fname] = pd.concat([vcg['VCGx'],vcg['VCGy'],vcg['VCGz']],axis=0,ignore_index=True)
```

```python
modes = 20
mode_obj = np.zeros([modes,len(pt_opt_vcg_theta.keys())])
mode_rms_norm = np.zeros([modes,len(pt_opt_vcg_theta.keys())])

for i,pt in enumerate(pt_opt_vcg_theta.keys()):
        for mode in range(1,21):
            mode_obj[mode-1,i] = calc_cost(df_vcg_vector(pt_opt_vcg_theta[pt].to_numpy()),df_vcg_vector(vcg_df['%s_m%i'%(pt,mode)].to_numpy()))
            mode_rms_norm[mode-1,i] = calc_cost_rms_norm(df_vcg_vector(pt_opt_vcg_theta[pt].to_numpy()),df_vcg_vector(vcg_df['%s_m%i'%(pt,mode)].to_numpy()))            
            
```

```python
opt_theta_all = [0.1072, 0.0466, 0.0352, 0.0742, 0.0542, 0.1182, 0.0507, 0.0715]
opt_theta_all = np.array(opt_theta_all)
opt_rms_all = [0.7074, 0.3250, 0.3925, 0.5576, 0.7007, 0.7588, 0.3989, 0.5570]
opt_rms_all = np.array(opt_rms_all)
```

```python
import seaborn as sns; sns.set_style("ticks")
fig, ax = plt.subplots(figsize = (5,5))
x_vals = np.arange(10)+1
for pt in range(8):
    plt.plot(x_vals, mode_obj[:10,pt], linewidth = 2)
plt.axhline(opt_theta_all.mean(),color='0.5',linestyle='--')
ax.set_xlim([0.5, 10.5])
#ax.set_ylim([0, 0.2])
ax.set_xlabel('Number of Modes',fontsize=16)
#ax.set_ylabel('Theta',fontsize=16)
#ax.set_title('Reconstructed rVCGs',fontsize=16)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
sns.despine()
#plt.show()
plt.savefig('Fig4B.pdf', transparent=True)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
x_vals = np.arange(10)+1
for pt in range(8):
    plt.plot(x_vals, mode_rms_norm[:10,pt], linewidth=2)
plt.axhline(opt_rms_all.mean(),color='0.5',linestyle='--')
ax.set_xlim([0.5, 10.5])
#ax.set_ylim([0, 0.2])
ax.set_xlabel('Number of Modes',fontsize=16)
ax.set_ylabel('RMS Error',fontsize=16)
#ax.set_title('Reconstructed Patient VCGs',fontsize=16)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
sns.despine()
#plt.show()
plt.savefig('Fig4C.pdf', transparent=True)
```
